🌟 Exercise 2: Descriptive Statistics
Task: Given a sample dataset, calculate the mean, median, variance, and standard deviation using SciPy.
Sample Dataset:
data = [12, 15, 13, 12, 18, 20, 22, 21]

```python
data = [12, 15, 13, 12, 18, 20, 22, 21]
```

```python
import numpy as np

scipy_describe = stats.describe(data)
mean_calc = scipy_describe.mean
median_calc = np.median(data)
variance_calc = scipy_describe.variance
std_dev_calc = np.std(data, ddof=1)  # Sample standard deviation
```

🌟 Exercise 3: Understanding Distributions
Task: Generate a normal distribution using SciPy with a mean of 50 and a standard deviation of 10. Plot this distribution.
Code Example:
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

```python
import matplotlib.pyplot as plt
from scipy.stats import norm

mean = 50
std_dev = 10

# Values spanning four standard deviations around the mean
x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 500)

# Calculate the probability density for each x-value
y = norm.pdf(x, loc=mean, scale=std_dev)

plt.plot(x, y)
plt.axvline(mean, color="red", linestyle="--", label="Mean = 50")

plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

🌟 Exercise 4: T-Test Application
Task: Perform a T-test on two sets of randomly generated data.
Code Example:
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

```python
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

t_statistic, p_value = stats.ttest_ind(data1, data2)
print(f"T-statistic: {t_statistic:.3f}")
print(f"P-value: {p_value:.3f}")
```

🌟 Exercise 5: Linear Regression Analysis
Objective: Apply linear regression to a dataset and interpret the results.

Task: Given a dataset of housing prices (house_prices) and their corresponding sizes (house_sizes), use linear regression to predict the price of a house given its size.
Dataset:
house_sizes: [50, 70, 80, 100, 120] (in square meters)
house_prices: [150,000, 200,000, 210,000, 250,000, 280,000] (in currency units)
Questions:
What is the slope and intercept of the regression line?
Predict the price of a house that is 90 square meters.
Interpret the meaning of the slope in the context of housing prices.

```python
house_sizes = [50, 70, 80, 100, 120] #square meters
house_prices = [150000, 200000, 210000, 250000, 280000] #currency units
x = house_sizes
y = house_prices
```

```python
#What is the slope and intercept of the regression line?
# Performing linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

# The slope and intercept of the regression line
print(f"Slope: {slope:.2f}")
print(f"Intercept: {intercept:.2f}")
```

```python
x = np.array(house_sizes)
y = np.array(house_prices)

plt.scatter(x, y, color="blue", label="Data points")
plt.plot(x, slope * x + intercept, color="red", label="Regression line")

plt.xlabel("House size (m²)")
plt.ylabel("House price")
plt.legend()
plt.show()
```

```python
#Predict the price of a house that is 90 square meters.
house_size = 90
predicted_price = slope * house_size + intercept

print(f"Predicted price for a 90 m² house: {predicted_price:,.2f}")
```

#Interpret the meaning of the slope in the context of housing prices.

The slope is the price of each additional square meter.

🌟 Exercise 6: Understanding ANOVA
Objective: Test understanding of ANOVA and its application.

Task: Three different fertilizers are applied to three separate groups of plants to test their effectiveness. The growth in centimeters is recorded.

Dataset:
fertilizer_1: [5, 6, 7, 6, 5]
fertilizer_2: [7, 8, 7, 9, 8]
fertilizer_3: [4, 5, 4, 3, 4]

Questions:
Perform an ANOVA test on the given data. What are the F-value and P-value?

Based on the P-value, do the fertilizers have significantly different effects on plant growth?

Explain what would happen if the P-value were greater than 0.05.

```python
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3= [4, 5, 4, 3, 4]
```

```python
#Perform an ANOVA test on the given data. What are the F-value and P-value?

# Performing one-way ANOVA
f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)
print(f"F-value: {f_value:.3f}")
print(f"P-value: {p_value:.3f}")
```

```python
#Based on the P-value, do the fertilizers have significantly different effects on plant growth?

if p_value < 0.05:
    print("The fertilizers have significantly different effects on plant growth (reject H0).")
```

#Explain what would happen if the P-value were greater than 0.05.

If the P-value were greater than 0.05, we would fail to reject the null hypothesis. This means there is not enough statistical evidence to conclude that the fertilizers have different effects on plant growth.

🌟 Exercise 7: Probability Distributions (Optional)
Work with a binomial distribution: calculate probabilities for different numbers of successes.

Example: Calculating the probability of getting exactly 5 heads in 10 coin flips.

```python
from scipy.stats import binom

# Parameters
n = 10  # number of trials (coin flips)
k = 5   # number of successes we want (heads)
p = 0.5 # probability of success on each trial (fair coin = 50%)

# Calculate probability of exactly 5 heads
prob_exactly_5 = binom.pmf(k, n, p)
print(f"Probability of exactly 5 heads: {prob_exactly_5:.4f}")

# You can also calculate cumulative probabilities
# Probability of 5 or fewer heads
prob_at_most_5 = binom.cdf(5, n, p)
print(f"Probability of at most 5 heads: {prob_at_most_5:.4f}")

# Probability of more than 5 heads
prob_more_than_5 = 1 - binom.cdf(5, n, p)
print(f"Probability of more than 5 heads: {prob_more_than_5:.4f}")
# Parameters
n = 10  # number of trials (coin flips)
k = 5   # number of successes we want (heads)
p = 0.5 # probability of success on each trial (fair coin = 50%)

# Calculate probability of exactly 5 heads
prob_exactly_5 = binom.pmf(k, n, p)
print(f"Probability of exactly 5 heads: {prob_exactly_5:.4f}")

# You can also calculate cumulative probabilities
# Probability of 5 or fewer heads
prob_at_most_5 = binom.cdf(5, n, p)
print(f"Probability of at most 5 heads: {prob_at_most_5:.4f}")

# Probability of more than 5 heads
prob_more_than_5 = 1 - binom.cdf(5, n, p)
print(f"Probability of more than 5 heads: {prob_more_than_5:.4f}")
```

🌟 Exercise 8: Correlation Coefficients (Optional)
Calculate the Pearson and Spearman correlation coefficients between two variables in a dataset.

Example in code: data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})

```python
import pandas as pd
data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})
```

```python
from scipy.stats import spearmanr, pearsonr

# Calculate Spearman Rank Correlation between Age and Income
corr, _ = spearmanr(data['age'], data['income'])
corr_pearson, _ = pearsonr(data['age'], data['income'])

print(f'Spearman Rank Correlation: {corr:.2f}')
print(f'Pearson Correlation: {corr_pearson:.2f}')
```

There is a perfect correlation between the age and income in this test dataset.
