1. Line Plot Exercise:

Create a line plot for the function y = x^2 for x ranging from 0 to 10. Add appropriate labels and a title.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
#y = x^2 for x ranging from 0 to 10
y = [x**2 for x in range(11)]
x = list(range(11))
```

```python
#Create a line plot for the function
plt.plot(x, y)
plt.title("Plot of y = x^2")
```

2. Bar Chart Exercise:

Visualize the sales of five products (e.g., Product A, Product B, etc.) in a bar chart. Customize the chart with different colors and patterns.

```python
#5 products
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
#Sales data for each product
sales = [150, 200, 300, 250, 400]
#bar chart for sales data, customize the chart with labels and title
plt.bar(products, sales)
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Sales Data for Each Product")
```

3. Histogram Exercise:

Generate a dataset of 500 random numbers following a normal distribution. Plot the histogram and experiment with different bin sizes and colors.

```python
#random dataset with 500 numbers following a normal distribution.
df = pd.DataFrame({'Random Numbers': [random.gauss(50, 15) for _ in range(500)]})
```

```python
#Plot the histogram and experiment with different bin sizes and colors.
plt.hist(df['Random Numbers'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel("Random Numbers")
plt.ylabel("Frequency")
plt.title("Histogram of Random Numbers")
```

4. Pie Chart Exercise:

Create a pie chart showing the percentage distribution of time spent on different daily activities (e.g., work, sleep, leisure, exercise, etc.). Use custom colors and add a shadow effect.

```python
#different daily activities (e.g., work, sleep, leisure, exercise, etc.)
activities = ['Work', 'Sleep', 'Leisure', 'Exercise', 'Other']
#time spent on each activity (in hours)
time_spent = [8, 7, 4, 2, 3]
#Create a pie chart showing the percentage distribution of time spent on different daily activities. Use custom colors and add a shadow effect.
plt.pie(time_spent, labels=activities, autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightblue', 'lightgreen', 'lightyellow'], shadow=True)
```
