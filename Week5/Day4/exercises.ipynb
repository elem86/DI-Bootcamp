🌟 Exercise 2: Creating a Line Plot for Temperature Variation

🌟 Exercise 4: Visualizing the Distribution of CGPA

Objective: Create a histogram to visualize the distribution of students’ CGPA.

Dataset Overview: Assume the CGPA data is categorized into ranges and loaded in a DataFrame named df.

Tasks:
Import necessary libraries.

Use Seaborn’s histplot to create a histogram of the CGPA categories.

Customize the histogram with a specific color and add a title.

Display the plot.

```python
df = pd.read_csv('Student Mental health.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
# Compute missing values and percentages per column
missing_values = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df)) * 100

missing_summary = pd.DataFrame({
    "Missing Values": missing_values,
    "Missing Percentage": missing_percentage
})

missing_summary = missing_summary[missing_summary["Missing Values"] > 0]

missing_summary
```

```python
# Fill missing values in the "Age" column with median "Age"
df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
#unique values in the CGPA column
df['What is your CGPA?'].unique()
```

```python
column = "What is your CGPA?"  # Replace with your exact column name

df[column] = df[column].str.strip()

category_order = [
    "0 - 1.99",
    "2.00 - 2.49",
    "2.50 - 2.99",
    "3.00 - 3.49",
    "3.50 - 4.00"
]

df[column] = pd.Categorical(
    df[column],
    categories=category_order,
    ordered=True
)

sns.histplot(
    data=df,
    x=column,
    discrete=True,
    shrink=0.8
)

plt.title("Distribution of CGPA Categories")
plt.xlabel("CGPA Range")
plt.ylabel("Number of Students")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

🌟 Exercise 5: Comparing Anxiety Levels Across Different Genders

Objective: Use a bar plot to compare the proportion of students experiencing anxiety across different genders.

Dataset Overview: The dataset includes columns: ‘Do you have Anxiety?’ and ‘Choose your gender’.

Tasks:

Import necessary libraries.

Use Seaborn to create a bar plot comparing anxiety levels across genders from the dataset df.

Customize the plot with an appropriate color palette and add a title.
Display the plot.

```python
anxiety_across_genders = (
    df["Do you have Anxiety?"].eq("Yes")
    .groupby(df["Choose your gender"])
    .mean()
    .rename_axis("Gender")
    .reset_index(name="Proportion with Anxiety")
)

plt.figure(figsize=(8, 5))
ax = sns.barplot(
    data=anxiety_across_genders,
    x="Gender",
    y="Proportion with Anxiety",
    hue="Gender",
    palette="Set2",
    legend=False
)
ax.set_ylim(0, 1)
ax.set_yticks([0, 0.2, 0.4, 0.6, 0.8, 1])
ax.set_yticklabels(["0%", "20%", "40%", "60%", "80%", "100%"])
plt.xlabel("Gender")
plt.ylabel("Proportion Reporting Anxiety")
plt.title("Proportion of Students Reporting Anxiety by Gender")
plt.tight_layout()
plt.show()
```

🌟 Exercise 6: Exploring the Relationship Between Age and Panic Attacks

Objective: 

Create a scatter plot to explore the relationship between students’ age and the occurrence of panic attacks.

Dataset Overview: The dataset includes columns: ‘Age’ and ‘Do you have Panic Attacks?’.

Tasks:

Import necessary libraries.

Convert panic attack responses to numeric values (e.g., Yes=1, No=0).

Use Seaborn’s scatterplot to create a scatter plot with ‘Age’ on the x-axis and numeric panic attack responses on the y-axis.

Customize the plot to improve readability by adding labels, a title, and adjusting point styles.

Display the plot.

```python
#Create a scatter plot to explore the relationship between students’ age and the occurrence of panic attacks.
#The dataset includes columns: ‘Age’ and ‘Do you have Panic Attack?’.
#Convert panic attack responses to numeric values (e.g., Yes=1, No=0).
panic = df.copy()
panic["Panic_Attack_Numeric"] = panic["Do you have Panic attack?"].map({
    "Yes": 1,
    "No": 0
})
panic = panic.dropna(subset=['Age', 'Do you have Panic attack?'])



```

```python
#Use Seaborn’s scatterplot to create a scatter plot with ‘Age’ on the x-axis and numeric panic attack responses on the y-axis.
# Add small random jitter to y-values for visualization only
panic["Panic_Attack_Jitter"] = panic["Panic_Attack_Numeric"] + np.random.uniform(
    -0.05, 0.05, size=len(panic)
)

# Scatter plot
plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=panic,
    x="Age",
    y="Panic_Attack_Jitter",
    hue="Do you have Panic attack?",
    s=80,
    alpha=0.7
)

plt.xlabel("Age")
plt.ylabel("Panic Attack Status")
plt.title("Relationship Between Age and Panic Attacks")
plt.yticks([0, 1], ["No", "Yes"])
plt.ylim(-0.2, 1.2)
plt.grid(True, alpha=0.3)

plt.show()
```
