```python
import numpy as np
from scipy import stats
```

```python
random_array = np.random.rand(1000) #floating point numbers
```

```python
mean_np = np.mean(random_array)
std_dev_np = np.std(random_array)

# SciPy uses numpy internally, so we use scipy.stats.describe()
scipy_describe = stats.describe(random_array)
mean_scipy = scipy_describe.mean
std_dev_scipy = np.sqrt(scipy_describe.variance)
```

```python
ex_time_np_mean = %timeit -o np.mean(random_array)
ex_time_scipy_mean = %timeit -o stats.describe(random_array).mean
ex_time_np_std = %timeit -o np.std(random_array)
ex_time_scipy_std = %timeit -o np.sqrt(stats.describe(random_array).variance)
```

```python
print("Execution time for np.mean:", ex_time_np_mean.best)
print("Execution time for scipy.stats.describe().mean:", ex_time_scipy_mean.best)
print("Execution time for np.std:", ex_time_np_std.best)
print("Execution time for scipy std:", ex_time_scipy_std.best)
```

```python
print(f"Mean comparison - NumPy: {mean_np}, SciPy: {mean_scipy}")
print(f"Std comparison - NumPy: {std_dev_np}, SciPy: {std_dev_scipy}")
print(f"\nNumPy mean is {ex_time_scipy_mean.best / ex_time_np_mean.best:.2f}x faster than scipy.stats.describe().mean")
print(f"NumPy std is {ex_time_scipy_std.best / ex_time_np_std.best:.2f}x faster than scipy std")
```
